```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("dataset/shopping.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# check the null values
df.isnull().sum()
```

```python
# fix the null values
df['Review Rating']=df.groupby('Category')['Review Rating'].transform (lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
# fix the lower and replace columns name
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(" ", "_")
```

```python
# can you change the purchase_amount_(usd)
df=df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
# df
```

```python
df.columns
```

```python
# df['purchase_amount']='$'+ df['purchase_amount'].astype(str)
# df.head()
```

```python
# Create a new columns age_group
labels=['Young-Adult','Adult','Mid-Age','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].sample(12)
```

```python
#create columns purachse_frequency_days

freq_mapping={
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-weekly':14,
    'Annually':365,
    'Every 3-month':90
}

df['purchase_frequency_days']=df['frequency_of_purchases'].map(freq_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
# check the applied for discount and promo_code_used
df[['discount_applied','promo_code_used']].head(10)
```

```python
# let's check promo code used 
(df['discount_applied']==df['promo_code_used']).all()
```

```python
df.head()
```

```python
# delete the promocode_used
df=df.drop('promo_code_used',axis=1)
```
